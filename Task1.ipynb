```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as pt
import seaborn as sns
```

**Import and Read the Dataset**

```python
url = "http://bit.ly/w-data"
df =  pd.read_csv(url)
df
```

**Print first five rows of data**

```python
df.head()
```

**To view details of the given datase**

```python
df.describe()
```

**Checking for null values**

```python
df.isnull().sum()
```

**Data Visualisation**

Plotting a graph to see the correlation between Hours and Scores


```python
df.plot(kind='scatter', x='Hours' , y = 'Scores')
pt.show()
```

```python
df.corr()
```

**Splitting the data into train and test data.**


*   Using the test_train_split()



```python
X_train , X_test , y_train , y_test = train_test_split( df.Hours, df.Scores , test_size = 0.2 , random_state = 0)
```

**Training the model**

```python
pt.scatter(X_train , y_train , label = 'Training_Data' , color='b')
pt.scatter(X_test , y_test , label = 'Testing_Data' , color='r')
pt.legend()
pt.title("Model Visualisation")
pt.show()
```

**Choosing the Linear Regression model for the computation and training the model** 

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

**Prediction of the trained model**

```python
pred = LR.predict(X_test.values.reshape(-1,1))
pred
```

**Using dataframe to compare the actual values and predicted values in a table**

```python
dfg = pd.DataFrame({'Actual values': y_test, 'Predicted values': pred})  
dfg
```

**Plotting a graph to show the actual and predicted values**

```python
graph = dfg.head()
graph.plot(kind='bar',figsize=(30,5))
pt.grid(which = 'both', color='red', linestyle='-.', linewidth=0.5)
pt.show()
```

**Plotting a graph for the test data**

```python
pt.plot(X_test , pred , label = 'LinearRegression' , color = 'b')
pt.scatter (X_test , y_test , label = 'Test_data' , color = 'r'  )
pt.legend()
pt.show()
```

**Using r2_score to check the accuracy of our predcition**

```python
accuracy = r2_score(y_test,pred)
accuracy
```

**Testing the model for "What will be predicted score if a student study for 9.25 hrs in a day?" as given in the Task**
Here the result is the predicted score after using the Linear Regression model.

```python
hour = 9.25
predict_score = LR.predict([[hour]])
print('Predicted Score for 9.25 hours is:', predict_score)
```
